```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from -2*pi to 2*pi
x = np.linspace(-2*np.pi, 2*np.pi, 400)
```

```python
y_sin = np.sin(x)

# Calculate the corresponding tangent values
y_tan = np.tan(x)

# Create a new figure
plt.figure()

# Plot the sine values
plt.plot(x, y_sin, label='sin(x)')

# Plot the tangent values
plt.plot(x, y_tan, label='tan(x)')

# Exclude large values to avoid asymptotes of the tangent function
plt.ylim(-10,10)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
y_cos = 1/np.cos(x)
# Calculate the corresponding tangent values
y_tan = 1/np.tan(x)

# Create a new figure
plt.figure()

# Plot the sine values
plt.plot(x, y_cos, label='sin(x)')

# Plot the tangent values
plt.plot(x, y_tan, label='tan(x)')

# Exclude large values to avoid asymptotes of the tangent function
plt.ylim(-10,10)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
